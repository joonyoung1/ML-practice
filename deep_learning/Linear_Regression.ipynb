```python
import matplotlib.pyplot as plt


X = [1, 2, 3, 4, 5, 6, 7]
Y = [25000, 55000, 75000, 110000, 128000, 155000, 180000]

plt.plot(X, Y)
plt.scatter(X, Y)

plt.show()
```

```python
def get_gradient(X, Y, w, b):
    w_gradient = 0
    b_gradient = 0
    for i in range(len(X)):
        w_gradient += (w * X[i] + b - Y[i]) * X[i]
        b_gradient += w * X[i] + b - Y[i]
    w_gradient *= 2 / len(X)
    b_gradient *= 2 / len(X)
    return w_gradient, b_gradient
```

```python
w = 0
b = 0
learning_rate = 0.001

for i in range(10000):
    w_gradient, b_gradient = get_gradient(X, Y, w, b)
    w -= w_gradient * learning_rate
    b -= b_gradient * learning_rate

w, b
```

```python
plt.scatter(X, Y)
plt.plot(range(1, 11), [w * x + b for x in range(1, 11)])

plt.show()
```
